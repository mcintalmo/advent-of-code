# Part I

```python
arr = np.array([list(line) for line in puzzle_input.strip().splitlines()])
df = pd.DataFrame(arr)
df = df.replace({"O": 1, ".": np.nan, "#": 0})
df
```

```python
cubes = df == 0
cubes
```

```python
gaps = cubes.cumsum()
gaps
```

```python
rocks_in_gap = {
    f"Col: {col}": df[col].groupby(gaps[col]).sum().to_dict()
    for i, col in enumerate(df.columns)
}
rocks_in_gap
```

```python
# Problem 1
height = df.shape[0]
weight = 0
for col in df.columns:
    rocks_in_gap = df[col].groupby(gaps[col]).sum()
    cubes_ind = (height - cubes[col][cubes[col]].index.to_numpy()).tolist()
    if not height in cubes_ind:
        cubes_ind = [height + 1] + cubes_ind
    rocks_in_gap.index = cubes_ind
    rocks_in_gap = rocks_in_gap.astype(int)
    for i, gap in rocks_in_gap.items():
        for rock in range(gap):
            weight += i - rock - 1
weight
```

# Part II

```python
puzz = pd.DataFrame(arr)
puzz = puzz.replace({"O": 1, ".": np.nan, "#": 0})
puzz
```

```python
new_df = puzz.copy().replace(1, np.nan)
new_df
```

```python
def gravity(df):
    new_df = df.copy().replace(1, np.nan)
    cubes = new_df == 0
    gaps = cubes.cumsum()
    for col in new_df:
        rocks_in_gap = df[col].groupby(gaps[col]).sum()
        cubes_ind = cubes[col][cubes[col]].index.to_numpy().tolist()
        if not 0 in cubes_ind:
            cubes_ind = [-1] + cubes_ind
        rocks_in_gap.index = cubes_ind
        rocks_in_gap = rocks_in_gap.astype(int)
        for i, gap in rocks_in_gap.items():
            for rock in range(gap):
                new_df[col][i + rock + 1] = 1
    return new_df


def rotate(df):
    return pd.DataFrame(np.rot90(np.rot90(np.rot90(df.copy()))))


def run_cycle(df, num_cycles=1):
    X = df.copy()
    for _ in range(4 * num_cycles):
        X = rotate(gravity(X))
    return X


def weigh(df):
    df = df.copy()
    df.index = df.shape[0] - df.index
    weight = (df.sum(axis=1) * df.index).sum()
    weight = int(weight)
    return weight
```

```python
state = puzz.copy()
run_cycle(state)
```

```python
run_cycle(state, 2)
```

```python
run_cycle(state, 3)
```

```python
past_states = np.zeros((1000, state.shape[0], state.shape[1]))
num_cycles = 1000000000
for i in range(num_cycles):
    state = run_cycle(state)
    cycle_np = state.fillna(-1).to_numpy()
    repeat = (cycle_np == past_states).all(axis=1).all(axis=1)
    if repeat.any():
        break
    past_states[i] = cycle_np

first_occurence = repeat.argmax()
period = i - first_occurence
remainder = (num_cycles - first_occurence) % period
print(
    f"First occurence: {first_occurence}\n"
    f"Period: {period}\n"
    f"Remainder: {remainder}"
)
```

```python
weigh(run_cycle(state, remainder - 1))
```
