```python
from pathlib import Path

import numpy as np
import pandas as pd
```

```python
input_data = Path("example.txt").read_text()
print(input_data)
```

## Part I

```python
lines = input_data.splitlines()
values = np.array([line.split() for line in lines[:-1]], dtype=int)
print(values)
```

```python
operations = np.array(lines[-1].split(), dtype=str)
add_mask = np.where(operations == "+", 1, 0).astype(bool)

add = values.T[add_mask].sum(axis=1)
prod = values.T[~add_mask].prod(axis=1)

result = add.sum() + prod.sum()
print(result)
```

## Part II

```python
values = [list(line) for line in lines[:-1]]
values = pd.DataFrame(values).sum().str.strip().add("   ").sum().split("    ")
max_len = max(len(v.split()) for v in values)
values = [
    [int(v) for v in value.split()] + [-1] * (max_len - len(value.split()))
    for value in values
    if value.strip()
]
values = np.array(values, dtype=int)
values = values.T
print(values)
```

```python
add = np.where(values > 0, values, 0).T[add_mask].sum(axis=1)
prod = np.where(values > 0, values, 1).T[~add_mask].prod(axis=1)

result = add.sum() + prod.sum()
print(result)
```
