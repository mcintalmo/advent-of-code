```python
from pathlib import Path

import numpy as np

input_data = Path("example.txt").read_text()
print(input_data)
```

```python
def is_between(
    x: np.ndarray,
    min_val: np.ndarray,
    max_val: np.ndarray,
) -> np.ndarray:
    return np.logical_and(
        x[:, np.newaxis] >= min_val[np.newaxis, :],
        x[:, np.newaxis] <= max_val[np.newaxis, :],
    )


range_string, ingredients_string = input_data.strip().split("\n\n")
range_lines = range_string.splitlines()
ranges = np.array(
    [list(line.split("-")) for line in range_lines],
    dtype=int,
)


id_lines = ingredients_string.splitlines()
ids = np.array(id_lines, dtype=int)


is_fresh = is_between(ids, ranges[:, 0], ranges[:, 1])
total_fresh = is_fresh.any(axis=1).sum()

print(total_fresh)
```

```python
sorted_ranges = np.sort(ranges, axis=0)
merged_ranges = []
current_start, current_end = sorted_ranges[0]
for start, end in sorted_ranges[1:]:
    if start <= current_end:
        current_end = max(current_end, end)
    else:
        merged_ranges.append((current_start, current_end))
        current_start, current_end = start, end
merged_ranges.append((current_start, current_end))
merged_ranges = np.array(merged_ranges, dtype=int)
all_fresh_ids = (merged_ranges[:, 1] - merged_ranges[:, 0] + 1).sum()
print(all_fresh_ids)
```
